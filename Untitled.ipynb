```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
categories = ['A', 'B', 'C', 'D']
values = [10, 25, 15, 30]
labels = ['Apples', 'Bananas', 'Oranges', 'Grapes']
sizes = [30, 20, 25, 25]
data = np.random.randn(1000)
```

```python
# Line plot
plt.figure(figsize=(8, 4))
plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Line Plot')
plt.legend()
plt.show()
```

```python
# Bar chart
plt.figure(figsize=(6, 4))
plt.bar(categories, values, color='blue', alpha=0.7)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart')
plt.show()
```

```python
# Scatter plot
plt.figure(figsize=(6, 4))
plt.scatter(x, y1, color='red', marker='o', label='sin(x)')
plt.scatter(x, y2, color='blue', marker='x', label='cos(x)')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot')
plt.legend()
plt.show()
```

```python
# Pie chart
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart')
plt.show()
```

```python
# Histogram
plt.figure(figsize=(8, 4))
plt.hist(data, bins=20, color='purple', alpha=0.7, edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```
